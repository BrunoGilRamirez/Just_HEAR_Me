# The mean of each input and Model's output
In the fallowing cells, it will be defined a function that calculate the mean of each output per sample. 

```python
def get_mean_tokens(df: pd.DataFrame) -> float:
    '''
    Description: get the mean of tokens with the LLaMA2 tokenizer
        df: a dataframe or a column of a dataframe. P/E df['column_name']
        return: the mean of tokens per sentence
    '''
    ntokens_each_sentence = []
    for i in range(len(df)):
        ntokens_each_sentence.append(len(tokenizer(df.iloc[i])['input_ids']))
        #round the mean withoout decimals
    return round(np.mean(ntokens_each_sentence))
```

```python
mean_tok_dict = {}
for model in df.columns:
    mean_tok_dict[model] = get_mean_tokens(df[model])
mean_tok_df = pd.DataFrame.from_dict(mean_tok_dict, orient='index', columns=['mean_tokens'])
#name of the index column
mean_tok_df.index.name = 'model'
mean_tok_df
```

# Frequency of unique words
in the following cell code, it will be defined functions to calculate frequency of unique words in a dataframe column and eliminate punctuation and stopwords from the dataframe column 

```python

def eliminate_punctuation(df: pd.DataFrame) -> pd.DataFrame:
    '''
    Description: eliminate punctuation from the dataframe
        df: a dataframe or a column of a dataframe. P/E df['column_name']
        return: a dataframe without punctuation
    '''
    for i in range(len(df)):
        #eliminate punctuation
        df.iloc[i] = df.iloc[i].replace(',', '')
        df.iloc[i] = df.iloc[i].replace('.', '')
        df.iloc[i] = df.iloc[i].replace('!', '')
        df.iloc[i] = df.iloc[i].replace('?', '')
        df.iloc[i] = df.iloc[i].replace('¿', '')
        df.iloc[i] = df.iloc[i].replace('¡', '')
        df.iloc[i] = df.iloc[i].replace("'", "")
        df.iloc[i] = df.iloc[i].replace('"', '')
        df.iloc[i] = df.iloc[i].replace(';', '')
        df.iloc[i] = df.iloc[i].replace(':', '')
        df.iloc[i] = df.iloc[i].replace('(', '')
        df.iloc[i] = df.iloc[i].replace(')', '')
        df.iloc[i] = df.iloc[i].replace('[', '')
        df.iloc[i] = df.iloc[i].replace(']', '')
        df.iloc[i] = df.iloc[i].replace('{', '')
        df.iloc[i] = df.iloc[i].replace('}', '')
        df.iloc[i] = df.iloc[i].replace('=', '')
        df.iloc[i] = df.iloc[i].replace('+', '')
        df.iloc[i] = df.iloc[i].replace('-', '')
        df.iloc[i] = df.iloc[i].replace('_', '')
        df.iloc[i] = df.iloc[i].replace('*', '')
        df.iloc[i] = df.iloc[i].replace('/', '')
        df.iloc[i] = df.iloc[i].replace('\\', '')
        df.iloc[i] = df.iloc[i].replace('|', '')
        df.iloc[i] = df.iloc[i].replace('@', '')
        df.iloc[i] = df.iloc[i].replace('#', '')
        df.iloc[i] = df.iloc[i].replace('$', '')
        df.iloc[i] = df.iloc[i].replace('%', '')
        df.iloc[i] = df.iloc[i].replace('^', '')
        df.iloc[i] = df.iloc[i].replace('&', '')
        df.iloc[i] = df.iloc[i].replace('~', '')
    return df

def get_unique_words(df: pd.DataFrame) -> set:
    '''
    Description: get a set of unique words from a dataframe column
        df: a dataframe or a column of a dataframe. P/E df['column_name']
        return: a set of unique words
    '''
    unique_words = set()
    newdf = eliminate_punctuation(df)
    for i in range(len(newdf)):
        # delete punctuation
        unique_words.update(newdf.iloc[i].lower().strip().split(' '))
    return unique_words, newdf

def get_frequency_words(df: pd.DataFrame) -> dict:
    '''
    Description: calculate frequency of unique words in a dataframe column
        df: a dataframe or a column of a dataframe. P/E df['column_name']
        return: a dictionary with unique words and their frequency
    '''
    unique_words,newdf = get_unique_words(df)
    frequency_words = {}
    for word in unique_words:
        frequency_words[word] = 0
    for i in range(len(newdf)):
        # delete punctuation
        
        for word in newdf.iloc[i].lower().strip().split(' '):
            frequency_words[word] += 1
    #add [',', '.', '!', '?', '¿', '¡', '', 'qué', 'cómo', 'cuándo', 'dónde', 'por', 'para', 'de', 'a', 'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'y', 'o', 'u', 'pero', 'sin', 'con', 'del', 'al', 'lo', 'le', 'les', 'me', 'te', 'se', 'nos', 'os', 'mi', 'tu', 'su', 'mío', 'tuyo', 'suyo', 'nuestro', 'vuestro', 'míos', 'tuyos', 'suyos', 'nuestros', 'vuestros', 'mía', 'tuya', 'suya', 'nuestra', 'vuestra', 'mías', 'tuyas', 'suyas', 'nuestras', 'vuestras', 'este', 'ese', 'aquel', 'esta', 'esa', 'aquella', 'estos', 'esos', 'aquellos', 'estas', 'esas', 'aquellas', 'este', 'esta', 'esto', 'ese', 'esa', 'eso', 'aquel', 'aquella', 'aquello', 'estos', 'estas', 'estas', 'esos', 'esas', 'esos', 'aquellos', 'aquellas', 'aquellas', 'este', 'esta', 'esto', 'ese', 'esa', 'eso', 'aquel', 'aquella', 'aquello', 'estos', 'estas', 'estas', 'esos', 'esas', 'esos', 'aquellos', 'aquellas', 'aquellas'] to stopwords
    stopwords.extend([',', '.', '!', '?', '¿', '¡', '','si', 'qué', 'cómo', 'cuándo', 'dónde', 'por', 'para', 'de', 'a', 'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'y', 'o', 'u', 'pero', 'sin', 'con', 'del', 'al', 'lo', 'le', 'les', 'me', 'te', 'se', 'nos', 'os', 'mi', 'tu', 'su', 'mío', 'tuyo', 'suyo', 'nuestro', 'vuestro', 'míos', 'tuyos', 'suyos', 'nuestros', 'vuestros', 'mía', 'tuya', 'suya', 'nuestra', 'vuestra', 'mías', 'tuyas', 'suyas', 'nuestras', 'vuestras', 'este', 'ese', 'aquel', 'esta', 'esa', 'aquella', 'estos', 'esos', 'aquellos', 'estas', 'esas', 'aquellas', 'este', 'esta', 'esto', 'ese', 'esa', 'eso', 'aquel', 'aquella', 'aquello', 'estos', 'estas', 'estas', 'esos', 'esas', 'esos', 'aquellos', 'aquellas', 'aquellas', 'este', 'esta', 'esto', 'ese', 'esa', 'eso', 'aquel', 'aquella', 'aquello', 'estos', 'estas', 'estas', 'esos', 'esas', 'esos', 'aquellos', 'aquellas', 'aquellas'])
    for word in stopwords:
        if (word in frequency_words):
            try:
                del frequency_words[word]
            except:
                pass
    frequency_words = dict(sorted(frequency_words.items(), key=lambda item: item[1], reverse=True))
    return frequency_words
```

# TF-IDF
In the following section, it will be calculated the importance of all terms in the 3 models output. Then will be ranked and compared how importance have changed between the 3 models. 

```python
# put all the elements from df column into a single text
def get_text(df: pd.DataFrame) -> str:
    '''
    Description: put all the elements from df column into a single text
        df: a dataframe or a column of a dataframe. P/E df['column_name']
        return: a string with all the elements
    '''
    #eliminate punctuation
    df = eliminate_punctuation(df)
    # eliminate stopwords
    for i in range(len(df)):
        df.iloc[i] = ' '.join([word for word in df.iloc[i].lower().strip().split(' ') if word not in stopwords])
    #eliminate numbers
    for i in range(len(df)):
        df.iloc[i] = ''.join([i for i in df.iloc[i] if not i.isdigit()])
    text = ''
    for i in range(len(df)):
        text += df.iloc[i] + ' '
    return text
```

```python
gpt3_output_text = get_text(df['GPT3_output'])
FTM_ouput_text = get_text(df['FTunned_Model_Output'])
BM_Output_text = get_text(df['Base_Model_Output'])
docs = [gpt3_output_text, FTM_ouput_text, BM_Output_text]
```

```python
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(docs) # tf_idf_vector contains the tf-idf scores of the words in the texts
tf_idf_array = tf_idf_vector.toarray() # tf_idf_array contains the tf-idf scores of the words in the texts
words_set = tr_idf_model.get_feature_names_out() # words_set contains all the words in the texts
order = np.argsort(tf_idf_array[0])[::-1] # order variable contains the indexes of the 50 most important words
# get the 50 most important words for each text with the scores
df_tfidf = pd.DataFrame({'GPT3_output': tf_idf_array[0][order], 'FTM_ouput': tf_idf_array[1][order], 'BM_Output': tf_idf_array[2][order]}, index=words_set[order]) # df_tfidf contains the tf-idf scores of the 50 most important words in the texts
df_tfidf.head(50)
```
