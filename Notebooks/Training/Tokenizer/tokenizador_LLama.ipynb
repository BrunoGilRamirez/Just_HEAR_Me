```python
data = datasets.DatasetDict({
    'train': datasets.Dataset.from_pandas(Emo_support_train)
})
data['train']
```

```python
CUTOFF_LEN = 1024

def generate_prompt(data_point):
#limita hasta 1024 tokens el texto
    text = data_point["texto"]#[:2460]
    return f"Intruccion: Analiza la Consulta, y determina la Emocion.\n{text}\n{data_point['etiqueta']}\n"

def get_training_corpus(train):
    return [train[i : i + 1000]for i in range(0, len(train), 1000)]

```

```python
train = []
for data_point in data['train']:
  train.append(generate_prompt(data_point))
```

```python
print(train[1:10])
```

```python
training_corpus = get_training_corpus(train)
```
